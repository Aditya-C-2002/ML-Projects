```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r"Diwali Sales Data.csv", encoding = "unicode_escape")
```

```python
df.shape #Check the number of rows and columns
```

```python
df.head(10) #Display the first 5 elements from the table
```

**Data Cleaning**

```python
df.info() # Provide the information
```

```python
df.drop(["Status", "unnamed1"], axis = 1, inplace = True)
# Use to drop the columns (because axis = 1 is columns)
```

```python
df.info()
```

```python
pd.isnull(df)
# USe to display ts there any Null Value in Table
```

```python
pd.isnull(df).sum()
# same as isnull but add the values
```

```python
df.dropna(inplace = True)
#dropna will drop NULL values
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.rename(columns = {'Marital_Status' : 'Married or Not'})
#not saved only display
```

**Exploratory Data Analysis**


**Gender**

```python
df.columns
```

```python
sns.countplot(x='Gender', data = df)
```

```python
gnd = sns.countplot(x = 'Gender', data = df)

for bars in gnd.containers:
    gnd.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sh = sns.barplot(x = 'Gender', 
                 y = 'Amount',
              data = sales_gen)

for bars in sh.containers:
    sh.bar_label(bars)
```

From Above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than man

**Age**

```python
df.columns
```

```python
sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
agegr = sns.countplot(data = df,
                         x = 'Age Group', 
                       hue ='Gender')

for bars in agegr.containers:
    agegr.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)

sns.barplot(x = 'Age Group', 
            y = 'Amount', 
         data = sales_age)
```

From above graph we can see that most of the buyers are age group between 26-35 years female

**State**

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x = 'State', y = 'Orders', data = sales_state)
```

```python
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x = 'State', 
            y = 'Amount', 
         data = sales_state)
```

From above graphs we can see that most of the orders & total sale/amount are from UP, Maharashtra and Karnaraka respectively

**Marital Status**

```python
ax = sns.countplot(data = df, x = "Marital_Status")
sns.set(rc={'figure.figsize':(1,1)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mar = df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc={'figure.figsize':(5,5)})
sns.barplot(x = 'Marital_Status', y = 'Amount', data = sales_mar, hue='Gender')
```

From above we can see that most of the buyers are married (women) and also having higher purchasing power

**Occupation**

```python
df.columns
```

```python
axiss = df["Occupation"].value_counts().index

plt.figure(figsize=(20,15))
ax = sns.countplot(data = df, 
                      x = 'Occupation', 
                  order = axiss, 
                    hue ='Occupation', 
                palette = None, 
                 legend = True)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_Occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

oc = sns.barplot(data = sales_Occ, 
                    x = 'Occupation', 
                    y = 'Amount', 
                  hue = 'Occupation',
              palette = None,
               legend = False)
```

From above graph we can see that most of the buyers are working in IT sector, HealthCare and Aviation

**Product Catagory**


```python
df.columns
```

```python
asiss = df["Product_Category"].value_counts().head(10).index

plt.figure(figsize=(40,15))
ax = sns.countplot(data = df, 
                      x = 'Product_Category', 
                  order = asiss)

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_pc = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(40,15))
sns.barplot(data = sales_pc, x='Product_Category', y='Amount')
```

From above graph we can see that most of the buyed products are Food, Clothing & Apparel and Electronics & Gadgets

**Most Sold Product**

```python
df.columns
```

```python
sales_pid = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(5)

# Plot the data
plt.figure(figsize=(30, 15))
sns.barplot(data = sales_pid, 
               x = "Product_ID", 
               y = "Orders", 
             hue = 'Orders', 
          legend = False, 
         palette = None)

# Add title and labels
plt.title('Top Products by Orders')
plt.xlabel('Product ID')
plt.ylabel('Total Orders')
#plt.xticks(rotation=45)  # Rotate x-tick labels for better readability
plt.show()
```

From the above graph we have seen that the top 5 selling producs are shown
